## Control Structures

- Sequential execution
  - Statements execute one after the other

- Transfer of control
  - you get to specify which statement to execute next, may not be sequential, i.e using if statements.

Sequence structure is built into Java, so statements will execute sequentially unless directed otherwise.

## Selection Structure

There are 3 types of selection statements:

1. `if` statements
1. `if ... else` statements
1. `switch` statements

### The `if/ if ... else` statment

Syntax:

```sh
if (condition) {
    //do thing
} else if (condition2) {
    //do another thing
} else {
    //better do this thing
}
```

**Note:** When nesting `if` statements, the `else` condition will evaulate against the innermost if.

### Ternary Operators

Are used to evaluate boolean expressions, and are declared as follows.

`<variable> var = (<boolean expression>) ? <if true> : <if false>`

### Switch Statement

Allows a variable to be tested for equality against mulitple cases.

Valid types of cases are: *int, byte, short, char, string*

Conclude each case with a `break` statement. If this is not added the program will continue to execute down the list of cases (irrespective of case evaluation), until it reaches a case that contains a `break` statement. This includes the `default` statement.
You may include a `default` case, which will execute if none of the cases evaulates true. This does not need to be concluded with a `break` statement.

Good practice for using switch statements is to have the default case at the end, and when declaring cases, ensure you test for all possible values.

## Repetition Structure

### Loops

See below for declaration.

- For: `for (<initialization statements>; <condition>; <iteration statements>) {<loop body>}`
- While: `while (<condition>) {<loop body>}`
- Do While: `do {<loop body>} while (<condition>);`

**Note:** `do ... while` loops will execute **at least** once, whereas `for` and `while` loops may not execute at all, depending on their condition evaluation.

#### Types Of Repetition Control

- Sentinal value: the program will continue to loop/ get input from the user until it is passed a predefined value that will cause the loop to stop.

- Counter controlled: the program will evaluate a variable at the start/ end of each loop, and only continue if it evaluates true. You can increment a counter within the program to achieve this.

#### Branching Statements

The `continue` statement will skip to the end of the innermost loop's body, before evaluating the boolean expression that controls the loop again. TL;DR: It skips all statements in the loop and starts from the top again if it still evaulates `true`.

> If you use `continue` in a while loop, ensure it does not skip over the incrementation of a control variable, as this is a common cause of an infinite loop.

The `break` statement will break out of whichever loop/ switch statement it is currently in, and continue executing sequentially as expected.

The `return` statement is used to delineate which variable to pass back to the method call, as well as exit the method itself. It does not need to be passed a variable to exit the method. The data type being returned from the method must match that of the method's declared value.

The `exit(n)` can be used to stop execution of the program. `n` is used to report the nature of the programs termination. Use 0 for success or non-zero for abnormal termination. There are values of `n` for specific types of termination, you can find these on the interweb.
